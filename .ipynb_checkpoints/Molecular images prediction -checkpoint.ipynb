### Duplicates removal

````python
# Duplicates removal from DF
og_size = len(df)
df.drop(df[df['Drug'].duplicated()].index, inplace = True)

print('Number of compounds removed:', og_size - len(df))
print('New shape after removing duplicates:',df.shape)
````

````python
# Obtain train, val, and test splits after cleaning data
def train_val_test_split(df, split):
    
    # Training
    train = df[df.columns][:len(split['train'])]
    
    # Validation
    val = df[df.columns][len(split['train']):(len(split['valid'])+len(split['train']))]
    
    # Test
    test = df[df.columns][(len(split['valid'])+len(split['train'])):len(df)]
    
    print('Data has been split')
    
    return train, val, test

train, val, test = train_val_test_split(df, split_AMES)
````

````python
# Convert the data from smile to molecule and convert to Image and save in file system in train, valid, test fold
# Example: Drug11.png

def convert_mol_to_image(data_split):
    for pos in range(len(data_split)):
        data_split[pos][1].reset_index(inplace = True)
        for ind in range(len(data_split[pos][1])):
            m = Chem.MolFromSmiles(data_split[pos][1]['Drug'][ind])
            name = data_split[pos][1]['Drug_ID'][ind].replace(" ", "")
            tox = data_split[pos][1]['Y'][ind] 
            
            # Create path if it does not exist
            path = "./images/"+data_split[pos][0]
            if not os.path.exists(path):
                os.makedirs(path)
            
            filename = path+"/%s%d.png" %(name, tox)
            img = Chem.Draw.MolToFile(m, filename)
````

````python
# Load and save all images 
convert_mol_to_image([('train', train), ('val', val), ('test', test)])
````

## Build datasets

In order to use DataLoader...

````python
class MolDataset(Dataset):
    def __init__(self, img_dir, transform = None):
        self.img_list = os.listdir(img_dir)
        self.img_labels = [int(filename.split('.')[0][-1]) for filename in self.img_list]
        self.root = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        # Path where the image is located 
        img_path = os.path.join(self.root, self.img_list[idx])
        # Image 
        image = read_image(img_path)
        #Label of the item
        label = self.img_labels[idx]
        # Transform
        if self.transform:
            image = self.transform(image)
        return image, label
````

We will make use of the auxiliary **display_image** function, given in HA1, to check if data has been built correctly

````python
def display_image(axis, image_tensor):
    """Display a tensor as an image
    
    Args:
        axis (pyplot axis)
        image_tensor (torch.Tensor): tensor with shape (num_channels=3, width, heigth)
    """
    
    # See hint above
    if not isinstance(image_tensor, torch.Tensor):
        raise TypeError("The `display_image` function expects a `torch.Tensor` " +
                        "use the `ToTensor` transformation to convert the images to tensors.")
        
    # The imshow commands expects a `numpy array` with shape (3, width, height)
    # We rearrange the dimensions with `permute` and then convert it to `numpy`
    image_data = image_tensor.permute(1, 2, 0).numpy()
    height, width, _ = image_data.shape
    axis.imshow(image_data)
    axis.set_xlim(0, width)
    # By convention when working with images, the origin is at the top left corner.
    # Therefore, we switch the order of the y limits.
    axis.set_ylim(height, 0)
````

## Building our baseline model

````python
# Map network output z to a hard label {0, 1}
def output_to_label(z):
    z[z >= 0.5] = 1
    z[z < 0.5] = 0
    return z.long()
````

````python
def training_loop(model, optimizer, lr_scheduler, loss_fn, train_loader, val_loader, num_epochs):
    print("Starting training")
    device = torch.device("cuda" if torch.cuda.is_available() 
                                  else "cpu")
    print(device)
    model.to(device)
    #Initialize train-val losses and accs vectors
    train_losses, train_accs, val_losses, val_accs = [], [], [], []
    
    # Iterate through number of vector indicated
    for epoch in range(1, num_epochs+1):
        # Check lr decay
#         for param_group in my_optim.param_groups:
#             print('Current LR:',param_group['lr'])
        model, train_loss, train_acc = train_epoch(model,
                                                   optimizer,
                                                   loss_fn,
                                                   train_loader,
                                                   val_loader,
                                                   device)
        # Compute val acc and loss after training
        val_loss, val_acc = validate(model, loss_fn, val_loader, device)
        print(f"Epoch {epoch}/{num_epochs}: "
              f"Train loss: {sum(train_loss)/len(train_loss):.3f}, "
              f"Train acc.: {sum(train_acc)/len(train_acc):.3f}, "
              f"Val. loss: {val_loss:.3f}, "
              f"Val. acc.: {val_acc:.3f}")
        
        #Add to vectors
        train_losses.extend(train_loss)
        train_accs.extend(train_acc)
        val_losses.append(val_loss)
        val_accs.append(val_acc)
        
        #Update LR (Decay)
        lr_scheduler.step()
    return model, train_losses, train_accs, val_losses, val_accs

def train_epoch(model, optimizer, loss_fn, train_loader, val_loader, device):
    # Train:
    model.train()
    train_loss_batches, train_acc_batches = [], []
    num_batches = len(train_loader)
    for batch_index, (x, y) in enumerate(train_loader, 1):
        inputs, labels = x.to(device), y.to(device)
        optimizer.zero_grad()
        z = model.forward(inputs)
        loss = loss_fn(z, labels.float())
        loss.backward()
        optimizer.step()
        train_loss_batches.append(loss.item())
        
        #predictions obtained in batch 
        hard_preds = output_to_label(z)
        #Accuracy obtained for batch
        acc_batch_avg = (hard_preds == labels).float().mean().item()
        train_acc_batches.append(acc_batch_avg)


    return model, train_loss_batches, train_acc_batches

def validate(model, loss_fn, val_loader, device):
    val_loss_cum = 0
    val_acc_cum = 0
    model.eval()
    with torch.no_grad():
        for batch_index, (x, y) in enumerate(val_loader, 1):
            inputs, labels = x.to(device), y.to(device)
            z = model.forward(inputs)
            batch_loss = loss_fn(z, labels.float())
            val_loss_cum += batch_loss.item()
            hard_preds = output_to_label(z)
            acc_batch_avg = (hard_preds == labels).float().mean().item()
            val_acc_cum += acc_batch_avg
    return val_loss_cum/len(val_loader), val_acc_cum/len(val_loader)
````

````python
class CNN1(nn.Module) :
    def __init__(self,img_size):
        super().__init__()
        self.img_size = img_size
        self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 10, kernel_size=3, stride=1, padding=0)
        self.conv2 = nn.Conv2d(in_channels = 10, out_channels = 10, kernel_size=3, stride=1, padding=0)
        self.fc1 = nn.Linear(((((self.img_size-3+1)//2 )-3+1)//2 )*((((self.img_size-3+1)//2 )-3+1)//2 )*10, 2)
        
        
    def forward(self, input_batch):
        self.input_batch = input_batch
        x = F.relu(self.conv1(self.input_batch))
        x = F.max_pool2d(x, kernel_size=2)
        
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, kernel_size=2)
        
        x = x.view(-1, ((((self.img_size-3+1)//2 )-3+1)//2 )*((((self.img_size-3+1)//2)-3+1)//2 )*10)
        x = self.fc1(x)
        x = F.softmax(x, dim=1)
        
        return x[:,1]
````

````python
# model
base_model = CNN1(128)
# Optimizer
my_optim = torch.optim.Adam(params = base_model.parameters(), lr=0.001)
# LR scheduler
my_lr_scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer=my_optim, gamma= 0.9)
````

````python
def train_model(m, my_optim, my_lr_scheduler, train_dataloader, val_dataloader):
    model, train_losses, train_accs, val_losses, val_accs = training_loop(model = m
                                                                        , optimizer= my_optim
                                                                        , lr_scheduler =my_lr_scheduler
                                                                        , loss_fn = nn.BCELoss()
                                                                        , train_loader = train_dataloader
                                                                        , val_loader = val_dataloader 
                                                                        , num_epochs = 60)
    return model, train_losses, train_accs, val_losses, val_accs
````

````python
model1, train_losses1, train_accs1, val_losses1, val_accs1 = train_model(base_model, my_optim, 
                                                                         my_lr_scheduler, train_dataloader, val_dataloader)
````

### Results visualization

````python
def plot_results(train_accs, train_losses, val_accs, val_losses):
    plt.figure(figsize=(7, 2), dpi = 95)

    sns.lineplot(x = numpy.arange(1,61)
                 , y = [round(sum(train_accs[i:(i+80)])/80, 3) for i in numpy.arange(0,len(train_losses), 80)], label = 'Training accuracy')
    sns.lineplot(x = numpy.arange(1,61), y = val_accs, label = 'Validation accuracy')
    plt.title('Accuracy for both sets')
    plt.xlabel('Number of epochs')
    plt.ylabel('Accuracy')
    plt.show()

    plt.figure(figsize=(7, 2), dpi = 95)
    sns.lineplot(x = numpy.arange(1,61)
                 , y = [round(sum(train_losses[i:(i+80)])/80, 3) for i in numpy.arange(0,len(train_losses), 80)], label = 'training loss')
    sns.lineplot(x = numpy.arange(1,61), y = val_losses, label = 'Validation loss')
    plt.title('Loss for both sets')
    plt.xlabel('Number of epochs')
    plt.ylabel('Loss')
    plt.show()
````

````python
plot_results(train_accs1, train_losses1, val_accs1, val_losses1)
````

While these are not bad metrics to start with, we note that we are not able to get above 0.75 in accuracy in our training set, which in turn prevents us from making progress with the validation set. It is possible that we are experiencing some overfitting in turn, so we need to increase complexity while introducing techniques that can reduce overfitting.

- Best metrics: **Epoch 44/60: Train loss: 0.543, Train acc.: 0.736, Val. loss: 0.641, Val. acc.: 0.633**

## Improving the model : CNN2

Same number of convolutional layers, but we increase the number of filters in the conv layers and we include dropout and L2 regularization to prevent possible overfitting

````python
class CNN2(nn.Module):
    def __init__(self, img_size):
        super().__init__()
        self.img_size = img_size
        
        # Layers
        self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 16, kernel_size = 3, stride=1, padding=0)
        self.conv2 = nn.Conv2d(in_channels = 16, out_channels = 16, kernel_size = 3, stride=1, padding=0)
        
        self.fc1 = nn.Linear(((((self.img_size-3+1)//2 )-3+1)//2 )*((((self.img_size-3+1)//2 )-3+1)//2 )*16, 2)

        
        self.drop = nn.Dropout(p = 0.5)
        
    def forward(self, input_batch):
        self.input_batch = input_batch
        
        #Conv layer 
        x = F.relu(self.conv1(self.input_batch))
        x = F.max_pool2d(x, kernel_size = 2)
        
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, kernel_size = 2)
        
        x = x.view(-1, ((((self.img_size-3+1)//2 )-3+1)//2 )*((((self.img_size-3+1)//2 )-3+1)//2 )*16)

        # Fully connected layer
        x = self.drop(x) 
        x = self.fc1(x)
        
        #Output layer
        x= F.softmax(x, dim=1)

        return x[:,1]
````

````python
# Model
cnn2 = CNN2(128)
# Optimizer
my_optim = torch.optim.Adam(params = cnn2.parameters(), lr=0.001, weight_decay = 1e-5)
# LR scheduler
my_lr_scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer = my_optim, gamma= 0.9)
````

````python
model2, train_losses2, train_accs2, val_losses2, val_accs2 = train_model(cnn2, my_optim, 
                                                                         my_lr_scheduler, train_dataloader, val_dataloader)
````

````python
plot_results(train_accs2, train_losses2, val_accs2, val_losses2)
````

By using two convolutional layers and a dropout at 0.5 prior to the fully connected layer, we are able to obtain promising metrics. In one of the epochs we obtain a validation accuracy of up to 0.67, while at no point do we exceed 0.75, indicating an organic evolution with little overfitting. The increase in filters has been a success, and the techniques introduced have allowed the model to generalise better, although we are still far from outstanding metrics.

- Best metrics: **Epoch 30/60: Train loss: 0.537, Train acc.: 0.734, Val. loss: 0.625, Val. acc.: 0.670**

## Improving the model : CNN3
We remove one convolutional layer to reduce complexity and see if better metrics can be achieved, i.e, trying to increase training accuracy hence validation as well. 

````python
class CNN3(nn.Module):
    def __init__(self, img_size):
        super().__init__()
        self.img_size = img_size
        
        # Layers
        self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 16, kernel_size = 3, stride=1, padding=0)
        self.fc1 = nn.Linear(((self.img_size-3+1)//2 )*((self.img_size-3+1)//2 )*16, 2)

        self.drop = nn.Dropout(p = 0.5)
        
    def forward(self, input_batch):
        self.input_batch = input_batch
        
        #Conv layer 
        x = F.relu(self.conv1(self.input_batch))
        x = F.max_pool2d(x, kernel_size = 2)
        
        x = x.view(-1, ((self.img_size-3+1)//2 )*((self.img_size-3+1)//2 )*16)

        # Fully connected layer
        x = self.drop(x) 
        x = self.fc1(x)
        
        #Output layer
        x= F.softmax(x, dim=1)

        return x[:,1]
````

````python
# Model
cnn3 = CNN3(128)
# Optimizer
my_optim = torch.optim.Adam(params = cnn3.parameters(), lr=0.001, weight_decay = 1e-5)
# LR scheduler
my_lr_scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer = my_optim, gamma= 0.9)
````

````python
model3, train_losses3, train_accs3, val_losses3, val_accs3 = train_model(cnn3, my_optim, 
                                                                         my_lr_scheduler, train_dataloader, val_dataloader)
````

````python
plot_results(train_accs3, train_losses3, val_accs3, val_losses3)
````

Although we achieved our goal of increasing accuracy in the training set, in the validation set we did not surpass the previous best metric we had obtained. The dropout and L2 regularisation may have contributed to reduce the overfitting, but it clearly still exists. We will opt to apply transformers in our next model to see if we can, by using a single convolutional layer, achieve better metrics.

- Best metrics obtained: **Epoch 34/60: Train loss: 0.315, Train acc.: 0.875, Val. loss: 0.723, Val. acc.: 0.657**

## Improving the model : CNN4
Finding a number of filters in between the two previous selections (10 and 16). Same dropout and L2 regularization techniques, but going back to non-transformed data. 

````python
class CNN4(nn.Module):
    def __init__(self, img_size):
        super().__init__()
        self.img_size = img_size
        
        # Layers
        self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 12, kernel_size = 3, stride=1, padding=0)
        self.fc1 = nn.Linear(((self.img_size-3+1)//2 )*((self.img_size-3+1)//2 )*12, 2)

        self.drop = nn.Dropout(p = 0.5)
        
    def forward(self, input_batch):
        self.input_batch = input_batch
        
        #Conv layer 
        x = F.relu(self.conv1(self.input_batch))
        x = F.max_pool2d(x, kernel_size = 2)
        
        x = x.view(-1, ((self.img_size-3+1)//2)*((self.img_size-3+1)//2)*12)

        # Fully connected layer
        x = self.drop(x) 
        x = self.fc1(x)
        
        #Output layer
        x= F.softmax(x, dim=1)

        return x[:,1]
````

````python
# Model
cnn4 = CNN4(128)
# Optimizer
my_optim = torch.optim.Adam(params = cnn4.parameters(), lr=0.001, weight_decay = 1e-5)
# LR scheduler
my_lr_scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer = my_optim, gamma= 0.9)
````

````python
model5, train_losses5, train_accs5, val_losses5, val_accs5 = train_model(cnn4, my_optim, 
                                                                         my_lr_scheduler, train_dataloader, val_dataloader)
````

````python
plot_results(train_accs5, train_losses5, val_accs5, val_losses5)
````

## Plot metrics obtained for each model

Lets compare all the models that have been developed to conclude our project

````python
def get_metrics_model(train_accs, val_accs):
    pos_best_val = numpy.argmax(val_accs)
    train = [round(sum(train_accs[i:(i+80)])/80, 3) for i in numpy.arange(0,len(train_accs), 80)][pos_best_val]
    val = round(val_accs[pos_best_val], 3)
    
    return train, val

metrics = []

#Append the metrics of every model
metrics.append(get_metrics_model(train_accs1, val_accs1))
metrics.append(get_metrics_model(train_accs2, val_accs2))
metrics.append((0.875, 0.657))
metrics.append(get_metrics_model(train_accs4, val_accs4))
metrics.append(get_metrics_model(train_accs5, val_accs5))
````

````python
def plot_comparison(names, met):
    sns.set(font_scale = 1)
    # set width of bar
    barWidth = 0.25
    plt.figure(figsize=(15, 6), dpi = 95)

    # set height of bar
    var1 = [i[0] for i in met]
    var2 = [i[1] for i in met]

    # Set position of bar on X axis
    br1 = numpy.arange(len(var1))
    br2 = [x + barWidth for x in br1]


    plt.bar(br1, var1, color ='dodgerblue', width = barWidth,
            edgecolor ='grey', label ='Train acc')
    plt.bar(br2, var2, color ='orange', width = barWidth,
            edgecolor ='grey', label ='Val acc')
    
    # Adding Xticks
    plt.xlabel('Models', fontweight ='bold', fontsize = 15)
    plt.ylabel('Metrics', fontweight ='bold', fontsize = 15)
    plt.xticks([r + barWidth for r in range(len(var1))],
            [name for name in names])
    plt.legend()
    plt.title('Metrics obtained for each CNN')
    
    plt.show()
    
plot_comparison(['CNN1','CNN2','CNN3','CNN4','CNN5'], metrics)
````

## Conclusions

- As can be seen, the best and most balanced results are obtained by using two convolutional layers, a number of filters around 16, and techniques to reduce overfitting, such as dropout and L2 regularisation. For our dataset in particular, the only transformations that made sense, randomrotation, do not allow us to generalise better, as the model does not learn correctly in the training set. We also note that the use of the learning rate decay technique allows us to advance at a moderately fast pace. 

- The use of images with the original resolution did not improve the results and increased the computational time required.

- Although we had hoped that the complexity reduction would lead to a strong improvement in our results, all it does is to trigger overfitting in the training set. By reducing the number of filters we wanted to find a middle ground but this was not possible. 

- Regarding transfer learning, the VGG-16 models usually bring strong improvements. However, the complexity is high, and we have seen with our designs that this is not indicative of an improvement. In addition, we doubt that it has been trained with such images, so freezing the weights is not a viable option. 

- Predicting molecular properties is a very complex task, and it is not always possible to obtain very high metrics using convolutional networks. With machine learning and the study of the physicochemical properties of each compound, the models are more accurate. 

- As future steps, we could consider the use of graph neural networks, mentioned in several articles, and other possible transformations that we may not have studied.
